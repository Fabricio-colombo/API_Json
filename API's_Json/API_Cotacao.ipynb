```python
#dentro do site (https://docs.awesomeapi.com.br/api-de-moedas), vamos pegar sua API grátis e fazer um teste.
#API >>> https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL

import requests
import json
```

```python
cotacoes = requests.get('https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL')
cotacoes_dic = cotacoes.json()
display(cotacoes_dic)
```

```python
#agora vou usar o get para pegar as informações de uma moeda específica e retornar o fechamento dos ultimos dias
#dentro do site ja tem >>> https://economia.awesomeapi.com.br/json/daily/:moeda/:numero_dias

cotacoes = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/10')
display(cotacoes.json())
```

```python
# https://economia.awesomeapi.com.br/:moeda/:quantidade
# Retorna cotações sequenciais de uma única moeda

cotacoes = requests.get('https://economia.awesomeapi.com.br/json/daily/EUR/5')
display(cotacoes.json())
```

```python
#agora vamos trazer a cotação de algumas moedas puxando pelo bid.
# cotacoes = requests.get('https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL')

cotacoes = requests.get('https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL')

print('Cotação do Dólar: ', cotacoes.json()['USDBRL']['bid'])
print('Cotação do Euro: ', cotacoes.json()['EURBRL']['bid'])
print('Cotação do Bitcoin: ', cotacoes.json()['BTCBRL']['bid'])
```

```python
#agora vou pegar a cotação do dolar, euro, bitcoin do inicio de 2023 até hoje.
#https://economia.awesomeapi.com.br/:moeda/:quantidade?start_date=20200301&end_date=20200330 >>> Retorna cotações sequenciais de um período específico

cotacoes_usd = requests.get('https://economia.awesomeapi.com.br/USD-BRL/360?start_date=20230101&end_date=20231231')
cotacoes_usd_dic = cotacoes_usd.json()

#criar um laço de repetição para pegar as cotações de cada dia e colocar em uma lista
#convertendo o dicionario em float

cotacoes_list = [float(item['bid']) for item in cotacoes_usd_dic]
print(cotacoes_list)
```

```python
#agora vamos usar o matplotlib para plotar um gráfico de linhas com as cotações do dolar
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.plot(cotacoes_list)
plt.title('Cotação do Dólar - 2023')
plt.xlabel('Dia')
plt.ylabel('Valor')
plt.show()

```
